# Homework:

Write a program to numerically evaluate the derivative of f(x)=cos x at an arbitrary position x.

1) Evaluate the derivative at position x using the forward difference scheme. Plot log(Error)=log10|f’(numerically)-f’(analytic)| against log10(dx) for dx=1e-16,1e-15,….,0.1 at x=π/4. You should find Error~$dx^n$; what is n? what is the optimal value for dx.

2) Repeat the exercise using the central difference scheme. Why this scheme is better?

3) Repeat (2) for x=0. Why do the results differ? (hint: write out the error term derived in the class explicitly by plugging the derivative of cos or sin)

```python
#Problem 1
import numpy as np
import matplotlib.pyplot as plt

#Forward difference from Lecture 4
def derifw(y,x,dx):
    return (y(x+dx)-y(x))/dx

x=np.pi / 4
dxarr=[]
yarr=[]
yexact = -np.sin(x) #exact analytic value from f'(x)
for dx in 10**np.arange(-16, -1, 1.):
    y=derifw(np.cos,x,dx)
    dxarr.append(np.log10(dx))
    yarr.append(np.log10(abs(y - yexact)))
    
plt.plot(dxarr, yarr)
plt.ylabel("log(Error)")
plt.xlabel("log(dx)")
plt.title("Forward Difference Error")
plt.show()
```

In the equation Error ~ dx^n, n = 2. The optimal value for dx is about 1e-8. 

```python
#Problem 2
import numpy as np
import matplotlib.pyplot as plt

#Central difference from Lecture 4
def dericd(y,x,dx):
    return (y(x+dx/2)-y(x-dx/2))/dx

x=np.pi / 4
dxarr=[]
yarr=[]
yexact = -np.sin(x) #exact analytic value from f'(x)
for dx in 10**np.arange(-16, 0, 1.):
    y=dericd(np.cos,x,dx)
    dxarr.append(np.log10(dx))
    yarr.append(np.log10(abs(y - yexact)))
    
plt.plot(dxarr, yarr)
plt.ylabel("log(Error)")
plt.xlabel("log(dx)")
plt.title("Central Difference Error, x = pi / 4")
plt.show()
```

This scheme is better because the values it uses to numerically calculate the derivative are better based around the actual values.
