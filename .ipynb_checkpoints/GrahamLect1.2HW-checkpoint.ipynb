```python
#Problem 1
import matplotlib.pylab as plt  # import Matplotlib label it as plt
import numpy as np   # import numpy library
from numpy import *

x=arange(-np.pi, np.pi, 0.1)
n=[11,21,31]

for i in range(len(n)):
    f=np.pi / 2
    for j in range(1, n[i]+2, 2):
        f=f+2*(sin(j*x)/j)
    plt.plot(x,f)

plt.legend(["n=11", "n=21", "n=31"])
plt.show()
```

```python
#Problem 2 - Sympy
from sympy import *
import matplotlib.pylab as plt
import numpy as np

#define symbols and constants
f, g = symbols('f g', cls=Function)
x = symbols('x')
n=11

f=np.pi / 2  #Set first value for f(x)
for j in range(1, n+2, 2):  #Add rest of f(x)
    f=f+2*(sin(j*x)/j)
    
# integrate f(x) to find g(x) and print
g = integrate(f, x)
print("g(x) = ", g)

#Now plot m(x) = g(x) - g(-pi) from -pi to pi
x1 = np.arange(-np.pi, np.pi, 0.1)
#y1 = g.subs(x, x1) - g.subs(x, -np.pi)
#y1 = np.array((len(x1),1))
y1 = []
for i in x1:
    value = g.subs(x, i) - g.subs(x, -np.pi)
    y1 = np.append(y1, value)

plt.plot(x1, y1)
```
