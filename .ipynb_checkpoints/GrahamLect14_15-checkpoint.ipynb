b) Merging compact objects produces a chirp signal in the gravitational wave detector.
Assume the chirp signal is 
\begin{equation}
y(t)=\begin{cases} sin(t^2) \quad 0<t<15 \\
0 \quad else\\
\end{cases}
\end{equation}

Please do wavelet transform to show the chirp signal is similar to the one you have seen
online (e.g.https://www.researchgate.net/figure/Spectrogram-of-signal-CHIRP-source-wwwligoorg_fig2_328932008). (Notice that the figure online has time as x-axis and frequency as the y-axis)
(You are welcome to use any wavelet you prefer.)

```python
#Problem A

#Nonlinear oscillator from Lecture 13 modified for our problem

import matplotlib.pylab as p
import numpy as np
import cmath                                            # complex math

#   Initialization
a = 0.
b = 7.
n = 500                                         
ydumb = np.zeros((2), float);    y = np.zeros((2), float)
fReturn = np.zeros((2), float);  k1 = np.zeros((2), float)
k2 = np.zeros((2), float);       k3 = np.zeros((2), float) 
k4 = np.zeros((2), float)
y[0] = 0.;   y[1] = -1.
t = a;       h = (b-a)/n;
xp=np.array([0.0])
yp=np.array([y[0]])

def f( t, y):                                           
    # Force function 
    fReturn[0] = y[1]                                             
    fReturn[1] = -0.01*y[0]**11  #modified from lecture 13
    return fReturn

def rk4(t,h,n):                 
    k1 = [0]*(n)
    k2 = [0]*(n)
    k3 = [0]*(n)
    k4 = [0]*(n)
    fR = [0]*(n)
    ydumb = [0]*(n)
    fR = f(t, y)                                            # Returns RHS's  
    for i in range(0, n):
        k1[i] = h*fR[i]                             
    for i in range(0, n):
        ydumb[i] = y[i] + k1[i]/2. 
    k2 = h*f(t+h/2., ydumb) 
    for i in range(0, n):
        ydumb[i] = y[i] + k2[i]/2. 
    k3 = h*f(t+h/2., ydumb)
    for i in range(0, n):
        ydumb[i] = y[i] + k3[i] 
    k4 = h*f(t+h, ydumb) 
    for i in range(0, 2):
        y[i] = y[i] + (k1[i] + 2.*(k2[i] + k3[i]) + k4[i])/6.
    return y    

while (t < b):                                              # Time loop
    if ((t + h) > b):
        h = b - t                                           # Last step
    y = rk4(t,h,2)
    t = t + h
    xp=np.append(xp, t)
    yp=np.append(yp,y[0])
    
p.plot(xp,yp)
p.title("Original Signal")
p.show()

#Decomposing into a Fourier series

N = 500;              Np = N         
twopi  = 2.*pi;       sq2pi = 1./sqrt(twopi);         h = twopi/N
dftz   = zeros( (Np), complex )                  # Complex elements   
nplot = zeros( N, float )   
fplot = zeros( N, float ) 

def fourier(dftz):                                              
    # DFT
    for n in range(0, Np):              
        zsum = complex(0.0, 0.0)                
        for  k in range(0, N):                              
            zexpo = complex(0, twopi*k*n/N)           # Complex exponent
            zsum += yp[k]*exp(-zexpo)
        dftz[n] = zsum * sq2pi      
        nplot[n] = n
        fplot[n] = sqrt(dftz[n].imag*dftz[n].imag+dftz[n].real*dftz[n].real)
        if abs(fplot[n]/fplot[0])>0.1: print("Contributes more than 10%: ", int(nplot[n]))


fourier(dftz)              # Call signal, transform
p.plot(nplot,fplot)
p.title("Fourier Transform")
p.show()

#Recompose original signal
invsignal = zeros( (N), float )
fplot2 = zeros( N, float ) 

def invfourier(invsignal):                                              # DFT
    for n in range(0, Np):              
        zsum = complex(0.0, 0.0)                
        for  k in range(0, N):                              
            zexpo = complex(0, twopi*k*n/N)           # Complex exponent
            zsum += dftz[k]*exp(zexpo)           
        invsignal[n] = zsum / sq2pi / N     
        fplot2[n] = np.sqrt(invsignal[n].imag*invsignal[n].imag+invsignal[n].real*invsignal[n].real)

invfourier(invsignal)
xa=xp[0:500]
p.plot(xa, invsignal.real)
p.plot(xp, yp)
p.title("Recomposed and Original Signal")
p.legend(("Recomposed", "Original"))
p.show()

```

```python
# Problem B

#Signal: y = sin(t^2) for 0 < t < 15 and 0 otherwise

import matplotlib.pyplot as plt
from numpy import *
import numpy as np

def func(t):
    return func1(t)

# sin wavelet within a time interval
def func1(t):
    if(0<t and t<15):
        return sin(t*t)
    else:
        return 0

step=0.1
t=arange(-100.,100.,step)

N=len(t)
y=zeros(N)
dftz=zeros(N,complex )
nplot=zeros(N)
fplot=zeros(N)
for idx in range(len(t)):
    y[idx]=func(t[idx])

fig = plt.figure()
plt.plot(t, y)
plt.xlim([-1., 20.])
plt.title("Original Signal")
plt.show()  

def fourier(signal):                                              # DFT
    for n in range(0, N):              
        zsum = complex(0.0, 0.0)                
        for  k in range(0, N):                              
            zexpo = complex(0, 2.*pi*k*n/N)           # Complex exponent
            zsum += signal[k]*exp(-zexpo)           
        dftz[n] = zsum * sqrt(2.*pi)      
        nplot[n] = n/N/step*2.*pi # the unit of \omega, cos(\omega t)
        fplot[n] = sqrt(dftz[n].imag*dftz[n].imag+dftz[n].real*dftz[n].real)
    
fourier(y)
fig = plt.figure() 
plt.plot(nplot,fplot)
plt.xlim([0,20.])
plt.title("Fourier Analysis")
plt.show()
```
