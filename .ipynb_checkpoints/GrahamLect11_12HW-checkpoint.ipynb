# Homework (last week's lecture notes):

Any mass can lead to a gravitational potential in 3-D (x-y-z) space
\begin{equation}
\nabla^2\Phi=4\pi G\rho
\end{equation}
    

We will set G=1 for this problem.  
Please solve the potential of a line mass 
(i.e. one zone at x=50, y=50 with nonzero density and all others with zero density) using Gauss-Siedel method. Since the problem has symmetry in the Z direction:
\begin{equation}
\nabla^2\Phi=\frac{\partial^2\Phi}{\partial^2 x}+\frac{\partial^2\Phi}{\partial^2y}
\end{equation}


The domain is X from 0 to 100, Y from 0 to 100. $\Delta$x= $\Delta$y=1. (assume $\Phi=0$
at the boundary and $\rho_{50,50}$ is always 1)

Please verify that the potential has radial dependence close to 2*G*$\mu$*ln(r)+constant. 
Where $\mu$ is its mass per unit length and r is the distance to the line mass. 



```python
import matplotlib.pylab as p;
from mpl_toolkits.mplot3d import Axes3D
from numpy import *;
import numpy;
print("Initializing")
Nmax = 100; Niter = 5000; V = zeros((Nmax, Nmax), float)   

V[50,50] = 4*numpy.pi # Line mass with non-zero density
    
for iter in range(Niter):                                  
    #if iter%10 == 0: print(iter)
    maxerr=0.
    for i in range(1, Nmax-1):                                                
        for j in range(1,Nmax-1): 
            du = 1./4.*(V[i+1,j]+V[i-1,j]+V[i,j+1]+V[i,j-1])-V[i,j] #
            if(maxerr<abs(du)): # how much correction we have in this step
                maxerr=abs(du)  
            V[i,j]= V[i,j]+du   
    #print(maxerr)
    if(maxerr<1.e-3):
        break
            
x = range(0, Nmax, 1)
y = range(0, Nmax, 1)                              
X, Y = p.meshgrid(x,y)                 

def functz(V):                                         # V(x, y) 
    z = V[X,Y]                        
    return z

Z = functz(V)                          
fig = p.figure()                                      # Create figure
ax = Axes3D(fig)                                      # Plot axes
ax.plot_wireframe(X, Y, Z, color = 'r')               # Red wireframe
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')
p.show()                                              # Show fig
```

```python
import matplotlib.pylab as p;
from mpl_toolkits.mplot3d import Axes3D
from numpy import *;
import numpy as np;

Nmax = 100; Niter = 5000; V = zeros((Nmax, Nmax), float)   

V[50,50] = 4*np.pi # Line mass with non-zero density
    
for iter in range(Niter):                                  
    #if iter%10 == 0: print(iter)
    maxerr=0.
    for i in range(1, Nmax-1):                                                
        for j in range(1,Nmax-1): 
            du = 1./4.*(V[i+1,j]+V[i-1,j]+V[i,j+1]+V[i,j-1])-V[i,j] #
            if(maxerr<abs(du)): # how much correction we have in this step
                maxerr=abs(du)  
            V[i,j]= V[i,j]+du   
    #print(maxerr)
    if(maxerr<1.e-3):
        break

x = range(0, Nmax, 1)
y = range(0, Nmax, 1)                              
X, Y = p.meshgrid(x,y)                 

def functz(V):                                         # V(x, y) 
    z = V[X,Y]                        
    return z

Z = functz(V)                          
fig = p.figure()                                      # Create figure
ax = Axes3D(fig)                                      # Plot axes
ax.plot_wireframe(X, Y, Z, color = 'r')               # Red wireframe
ax.set_xlabel('X')                                     
ax.set_ylabel('Y')
ax.set_zlabel('Potential')

#Set up second mesh with radial dependence function
V2 = zeros((Nmax, Nmax), float)

for i in range(1, Nmax-1):                                                
    for j in range(1,Nmax-1):
        mu = (V[i,j]/4/np.pi)**(1/3)
        r = np.sqrt((50-i)**2 + (50-j)**2)
        V2[i,j] = 2 * mu * np.log(r)/10
        
Z2 = functz(V2)
ax.plot_wireframe(X, Y, Z2, color='b')

p.show()                                              # Show fig
```
